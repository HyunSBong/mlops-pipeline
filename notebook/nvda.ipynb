#merging datasets

```python
import pandas as pd

# Load the uploaded CSV files
sentiment_df = pd.read_csv('/content/nvda_sentiment_train.csv')
close_df = pd.read_csv('/content/nvda_close_mar_to_may_2025.csv')


```

```python
# Remove the first row from the close_df which contains redundant header-like data
close_df_cleaned = close_df.drop(index=0).reset_index(drop=True)

# Ensure the 'Date' columns are in datetime format for accurate merging
sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
close_df_cleaned['Date'] = pd.to_datetime(close_df_cleaned['Date'])

# Merge the dataframes on 'Date'
merged_df = pd.merge(sentiment_df, close_df_cleaned, on='Date', how='inner')


```
